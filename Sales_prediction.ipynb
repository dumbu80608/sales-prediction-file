```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = pd.read_csv("/content/sample_data/advertising.csv")
print(data.head())
print(data.isnull().sum())
```

```python
import plotly.express as px
import plotly.graph_objects as go
fi= px.scatter(data_frame = data, x="Sales",
                    y="TV", size="TV", trendline="ols")
fi.show()
```

```python
fi=px.scatter(data_frame = data, x="Sales",
                    y="Newspaper", size="Newspaper", trendline="ols")
fi.show()
```

```python
fi=px.scatter(data_frame=data,x="Newspaper",y="Sales",trendline="ols")
fi.show()
```

```python
x = np.array(data.drop(["Sales"],axis= 1))
y = np.array(data["Sales"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y,
                                                test_size=0.2,
                                                random_state=42)

m=LinearRegression()
m.fit(xtrain, ytrain)
print(m.score(xtest, ytest))

#f=[[TV, Radio, Newspaper]]
f=np.array([[230.1, 37.8, 69.2]])
print(model.predict(f))
```
